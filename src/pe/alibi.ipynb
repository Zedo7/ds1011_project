ALiBi: head-specific linear distance biases.


```python

import torch, math
def alibi_slopes(n_heads):
    m = 2 ** torch.arange(0, math.ceil(math.log2(n_heads))).float()
    return (1.0 / m)[:n_heads]
def alibi_bias(L, n_heads):
    i = torch.arange(L).view(1,1,L,1); j = torch.arange(L).view(1,1,1,L)
    dist = j - i
    slopes = alibi_slopes(n_heads).view(1,n_heads,1,1)
    return -slopes * torch.relu(dist.float())
```
