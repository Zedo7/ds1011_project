RoPE: rotate Q/K by position-dependent angles.


```python

import torch
def rope_angles(L, d):
    inv = 1.0 / (10000 ** (torch.arange(0, d, 2).float() / d))
    t = torch.arange(L).float().unsqueeze(1)
    return t * inv.unsqueeze(0)  # (L, d/2)
def apply_rope(x, theta):
    x1 = x[...,0::2]; x2 = x[...,1::2]
    cos, sin = theta.cos(), theta.sin()
    xr = torch.stack([x1*cos - x2*sin, x1*sin + x2*cos], dim=-1).flatten(-2)
    return xr
# integrate by applying to q/k before attention dot-products
```
